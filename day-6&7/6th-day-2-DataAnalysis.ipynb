```python
df = pd.read_csv("data/yelp.csv")
df.head()   
```

```python
df.shape
```

### Lowercase

```python
df.text = df.text.str.lower()
df.head()
```

### Punctuation

```python
df.text = df.text.str.replace(r'[^\w\s]', '')
df.head()
```

### Numbers

```python
df.text = df.text.str.replace(r'\d+', '')
df.head()
```

### New line and tab characters

```python
df.text = df.text.str.replace(r'\n+', ' ')
df.text = df.text.str.replace(r'\r+', ' ')
df.head()
```

```python
df.text[0]
```

```python
ybw =df[(df['stars'] == 1) | (df['stars'] == 5)]
ybw.reset_index(drop=True, inplace=True)
ybw.head()
```

```python
x = ybw[['text']]
y = ybw['stars']
from sklearn.model_selection import train_test_split
vect = CountVectorizer(lowercase=True, stop_words='english',ngram_range=(1,2))
x_new = vect.fit_transform(x.text)
x_train, x_test, y_train, y_test = train_test_split(x_new, y, test_size=0.2, random_state=42)

from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(x_train, y_train)

y_pred = nb.predict(x_test)

from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

## Get data from X(twitter) and cleaning

```python
# pip install ntscraper
from ntscraper import Nitter
ts = Nitter()
```

```python
# tweets = ts.get_tweets("elonmusk", mode= "user", number= 100)
tweets2 = ts.get_tweets("Fatihaltayli", mode= "user", number= 50)
```

```python
df_tweets = pd.DataFrame(tweets2["tweets"])
```

```python
df_tweets.head()
```

```python
df_tweets["name"] = df_tweets["user"].apply(lambda x: x["name"])
df_tweets["username"] = df_tweets["user"].apply(lambda x: x["username"])
```

```python
df_tweets.head()
```

```python
df_tweets["comments"] = df_tweets["stats"].apply(lambda x: x["comments"])
df_tweets["likes"] = df_tweets["stats"].apply(lambda x: x["likes"])
df_tweets.head()
```

```python
df_tweets["text"] = df_tweets["text"].apply(nfx.clean_text)
df_tweets["text"] = df_tweets["text"].apply(nfx.remove_special_characters)
df_tweets["text"] = df_tweets["text"].apply(nfx.remove_stopwords)
df_tweets["text"] = df_tweets["text"].apply(nfx.remove_multiple_spaces)
df_tweets["text"] = df_tweets["text"].apply(nfx.remove_urls)
df_tweets["text"] = df_tweets["text"].apply(nfx.remove_puncts)
df_tweets["text"] = df_tweets["text"].apply(nfx.remove_stopwords)

df_tweets.text[0]
```
