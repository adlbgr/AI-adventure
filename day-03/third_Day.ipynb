### Feature engineering

```python
df.corr(numeric_only=True)
```

```python
abs(df.corr(numeric_only=True)['Price'].sort_values(ascending=False))
```

```python
x=df[['Mileage','Cylinder','Liter','Cruise']]
y=df[['Price']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
# 42 otostopçunun galaksi rehberi adli kitaptan alıntı bir sayı
```

```python
x_train.shape
```

## Modelling

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
model=lr.fit(x_train,y_train) # makine öğrenmesi
```

```python
tahmin=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
mean_squared_error(y_test,tahmin)**.5
```

```python
r2_score(y_test,tahmin)
```

```python
#Label Encoding
```

```python
x=df[['Mileage','Cylinder','Liter','Cruise','Make','Trim']]
y=df[['Price']]
```

```python
x.head()
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
model=lr.fit(x_train,y_train)
```

```python
tahmin=model.predict(x_test)
```

```python
mean_squared_error(y_test,tahmin)**.5 #Root Mean Squared error hata
```

```python
r2_score(y_test,tahmin)
```

```python
from sklearn.linear_model import Ridge,Lasso,ElasticNet
```

```python
R=Ridge()
L=Lasso()
E=ElasticNet()
```

```python
Rmodel=R.fit(x_train,y_train)
```

```python
tahmin2=Rmodel.predict(x_test)
```

```python
mean_squared_error(y_test,tahmin2)**.5
```

```python
r2_score(y_test,tahmin2)
```

```python
Lmodel=L.fit(x_train,y_train)
```

```python
tahmin3=Lmodel.predict(x_test)
```

```python
mean_squared_error(y_test,tahmin3)**0.5
```

```python
r2_score(y_test,tahmin3)
```

```python
Emodel=E.fit(x_train,y_train)
```

```python
tahmin4=Emodel.predict(x_test)
```

```python
mean_squared_error(y_test,tahmin4)**0.5
```

```python
#CCoefficients-Katsayılar
```
