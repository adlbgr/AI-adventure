```python
ratings = pd.merge(ratings, movies)
ratings
```

```python
ratings.title.value_counts()
```

```python
ratings.movie_id.nunique()
```

```python
movie_ratings = ratings.pivot_table(index=['user_id'], columns=['title'], values='rating')
```

```python
movie_ratings
```

```python
movie_ratings["Star Wars (1977)"]
```

```python
starwars_ratings = movie_ratings["Star Wars (1977)"]
starwars_ratings.value_counts(), starwars_ratings.mean()
```

```python
movie_ratings[['Star Wars (1977)', 'Empire Strikes Back, The (1980)']].corr(numeric_only=True)
```

```python
movie_ratings.corrwith(starwars_ratings).sort_values(ascending=False)
```

```python
similar_movies = movie_ratings.corrwith(starwars_ratings)
similar_movies = similar_movies.dropna()
similar_movies.sort_values(ascending=False)
```

```python
movie_stats = ratings.groupby('title').agg({'rating': [np.size, np.mean]})
movie_stats.head()
```

```python
popouler_movies = movie_stats[movie_stats['rating']['size'] >= 100]
popouler_movies = popouler_movies.sort_values([('rating', 'mean')], ascending=False)
popouler_movies.head(),popouler_movies.shape
```

```python
popouler_movies.columns = popouler_movies.columns.get_level_values(0)
df = popouler_movies.join(pd.DataFrame(similar_movies, columns=['similarity']))
```
