# Scaling-Normalization- PCA

```python
import pandas as pd
df = pd.read_csv('data/pima-indians-diabetes.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
from sklearn.preprocessing import normalize,scale
x = df.drop('Outcome',axis=1)
y = df['Outcome']
```

## Scale

```python
x = scale(x)
x = pd.DataFrame(x,columns=df.columns[:-1])
```

```python
x.head()
```

```python
x.describe().T
```

## Normalized

```python
norm_x = normalize(x)
norm_x = pd.DataFrame(norm_x,columns=df.columns[:-1])
norm_x.head()
```

## Boyut Küçültme PCA - Principle Component Analysis

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_x = pca.fit_transform(norm_x)
pca_x = pd.DataFrame(pca_x,columns=['PC1','PC2'])
pca_x.head()
```
