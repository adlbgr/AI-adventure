#### Complex Queries

13. Find the subset of `df` where the clients are females older than 30 years.
14. Repeat problem 1, but return only the hair color and eye color.
15. Find the unique combination of hair and eye color for women older than 25 years.



```python
df_females_over_30 = df[(df['GENDER'] == 'F') & (df['AGE'] > 30)]
df_females_over_30

```

```python
women_over_25 = df[(df['GENDER'] == 'F') & (df['AGE'] > 25)]
unique_combinations = women_over_25.groupby(['HAIR_COLOR', 'EYE_COLOR']).size().reset_index(name='COUNT')
unique_combinations
```
